# 1.Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import association_rules,apriori
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

# 2.Import Dataset

```python
books=pd.read_csv('book.csv')
books.head()
```

# 3.Data Understanding | EDA

```python
books.shape
```

```python
books.info()
```

```python
books.isna().sum()
```

# Apriori 

## support=0.1 and confidence=0.5

```python
frequency=apriori(df=books,min_support=0.1,use_colnames=True,)
```

```python
frequency
```

```python
frequency['length']=frequency['itemsets'].apply(lambda x:len(x))
frequency
```

```python
best_associates=association_rules(df=frequency,metric='lift',min_threshold=0.5)
best_associates
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x=best_associates['support'],y=best_associates['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```
